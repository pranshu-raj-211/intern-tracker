```python
threshold = 10  # Set your threshold here
counts = data['location'].value_counts()
data['location'] = data['location'].apply(lambda x: x if counts[x] > threshold else 'Other')
data['location'].value_counts(sort=True)
```

```python
def process_location(location):
    if location in ['Full-time', 'Internship', 'Contract']:
        return 'Other'
    elif 'remote' in location.lower():
        return 'remote'
    else:
        parts = location.split(',')
        if len(parts) >= 2:
            return parts[-2].strip()
        else:
            return 'Other'

data['location'] = data['location'].apply(process_location)
```

```python
data['location']=data['location'].apply(lambda x:x.lower())
data['location'].value_counts(sort=True)
```
